```python
import plotly.express as px
```

```python
#Definir el formato del título
def format_title(title, subtitle=None, subtitle_font_size=14):
        title = f'<b>{title}</b>'
        if not subtitle:
            return title
        subtitle = f'<span style="font-size: {subtitle_font_size}px;">{subtitle}</span>'
        return f'{title}<br>{subtitle}'
```

```python
fig = px.bar(df_prom , x = 'genre', y = 'avg_vote')
fig.update_layout(
  yaxis_title = 'Promedio',
  xaxis_title = 'Género',
  title_text = format_title('Géneros de películas', 'Promedio'),
  template = 'plotly_white'
)
fig
```

```python
df_num_pel = df.groupby(['year', 'genre']).size().reset_index(name='Number_of_Movies')
df_num_pel
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['year']
  ys = series['Number_of_Movies']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df_num_pel.sort_values('year', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('year')
_ = plt.ylabel('Number_of_Movies')
```
