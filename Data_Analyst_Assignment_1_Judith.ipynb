```python
# YOUR CODE HERE:
from datetime import datetime, date
import plotly.express as px

a_id = df.asset_id.unique()
case_data = df[df['asset_id'] == a_id[5]]
case_data['date'] = pd.to_datetime(case_data['time_est']).dt.date
case_data.time_est.nunique()
case_data = case_data.groupby(['Site', 'Zone']).first().reset_index()
case_data = case_data.sort_values(by = 'date', ascending=True).reset_index()

#Definir el formato del título
def format_title(title, subtitle=None, subtitle_font_size=14):
        title = f'<b>{title}</b>'
        if not subtitle:
            return title
        subtitle = f'<span style="font-size: {subtitle_font_size}px;">{subtitle}</span>'
        return f'{title}<br>{subtitle}'

from string import hexdigits
fig = px.line(case_data, x='Zone', y='date', height=800, color = 'Site', markers = True)
fig.update_layout(
    yaxis_title='Time',
    xaxis_title='Site / POI',
    title_text='Location of the case over time',
    template='plotly_white'
)
fig
```

### Question 6:

The customer wants to understand the efficieny of ther DC operations.
* Based on what you see in the data, (all zones except for STORE), which parts of their operation are most & least "efficient? (10 pts)

```python
# YOUR CODE HERE
data = df.copy()
data['time_est'] = pd.to_datetime(data['time_est'])

# Filtrar los datos para excluir la zona de la tienda
data_dc = data[data['Site'] != 'Store']

# Calcular el tiempo que los activos pasan en cada zona
data_dc['Time_in_zone'] = data_dc.groupby(['asset_id', 'Zone'])['time_est'].diff().fillna(pd.Timedelta(seconds=0))

# Agrupar por zona y calcular el tiempo promedio en cada zona
average_time_in_zone = data_dc.groupby('Site')['Time_in_zone'].mean()

# Mostrar los resultados
print("Tiempo promedio en cada zona:")
print(average_time_in_zone)
```
