```python
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
#TRATAMIENTO DE DATOS

#Cambiar las variables tipo categoricas por variables tipo numericas

df_corr = df.copy()
df_corr['Sex'] = df_corr['Sex'].replace({'Female':0,'Male':1})

df_corr['AgeCategory'] = df_corr['AgeCategory'].replace({
    '18-24': 0,
    '25-29': 1,
    '30-34': 2,
    '35-39': 3,
    '40-44': 4,
    '45-49': 5,
    '50-54': 6,
    '55-59': 7,
    '60-64': 8,
    '65-69': 9,
    '70-74': 10,
    '75-79': 11,
    '80 or older': 12
})

df_corr['HadDiabetes'] = df_corr['HadDiabetes'].replace({'No':0,
                                         'No, pre-diabetes or borderline diabetes':0,
                                         'Yes':1,
                                         'Yes, but only during pregnancy (female)':1})


df_corr['GeneralHealth'] = df_corr['GeneralHealth'].replace({'Poor': 0,
                                                     'Fair': 1,
                                                     'Good': 2,
                                                     'Very good': 3,
                                                     'Excellent': 4})
df_corr['CovidPos'] = df_corr['CovidPos'].replace({'Tested positive using home test without a health professional': 1})
df_corr['SmokerStatus'] = df_corr['SmokerStatus'].replace({'Former smoker': 1,
                                                           'Never smoked': 0,
                                                           'Current smoker - now smokes every day': 1,
                                                         'Current smoker - now smokes some days':1})

#Cambia todas las variables que tengan como respuesta 'Yes'/'No' por 0/1 respectivamente.
df_corr = df_corr.replace('No',0)
df_corr = df_corr.replace('Yes',1)


#Eliminación de columnas
df_corr = df_corr.drop(columns = 'RaceEthnicityCategory')
df_corr = df_corr.drop(columns = 'TetanusLast10Tdap')
df_corr = df_corr.drop(columns = 'ECigaretteUsage')
df_corr = df_corr.drop(columns = 'RemovedTeeth')
df_corr = df_corr.drop(columns = 'LastCheckupTime')
df_corr = df_corr.drop(columns = 'State')
df_corr = df_corr.drop(columns = 'ChestScan')
df_corr = df_corr.drop(columns = 'HIVTesting')
df_corr = df_corr.drop(columns = 'HighRiskLastYear')
#Imprimir el dataset
df_corr
```

```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
```

```python
def plot_confusion_matrix(cm, labels):
    fig_cm = px.imshow(cm, labels=dict(x="Predicted", y="Actual", color="Count"),
                       x=labels, y=labels, color_continuous_scale='Blues', text_auto = True,
                       title="Confusion Matrix")
    fig_cm.update_layout(coloraxis_showscale=False)
    fig_cm.show()
```

```python
df_mod = df_corr.copy()
```

```python
df_mod.shape
```

```python
X = df_mod.drop(columns=['HadHeartAttack'])
y = df_mod['HadHeartAttack']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

##Logistic Regression

**Train Model**

```python
lr = LogisticRegression()
lr.fit(X_train_scaled, y_train)
```

**Evaluate Model**

```python
y_pred = lr.predict(X_test_scaled)
y_pred_proba = lr.predict_proba(X_test_scaled)[:, 1]
```

```python
lr_accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {lr_accuracy:.2f}")
```

```python
lr_precision = precision_score(y_test, y_pred)
print(f"Precision: {lr_precision:.2f}")
```

```python
lr_recall = recall_score(y_test, y_pred)
print(f"Recall: {lr_recall:.2f}")
```

```python
lr_f1 = f1_score(y_test, y_pred)
print(f"F1 Score: {lr_f1:.2f}")
```

```python
report = classification_report(y_test, y_pred)
print("Classification Report:")
print(report)
```

```python
lr_cm = confusion_matrix(y_test, y_pred)
plot_confusion_matrix(lr_cm, ['No tener un ataque al corazón', 'Tener un ataque al corazón'])
```

```python
lr_cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(lr_cm, annot=True, cmap='Spectral', fmt='g', xticklabels=['No tener un ataque', 'Tener un ataque'], yticklabels=['No tener un ataque', 'Tener un ataque'])
```
